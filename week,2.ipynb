```python
# Intermediate Challenge##Write a function to calculate mean and median of a list
def calculate_mean_median(numbers):
    if not numbers:
        return None, None

    n = len(numbers)
    numbers_sorted = sorted(numbers)

    # Mean
    mean = sum(numbers) / n

    # Median
    if n % 2 == 0:
        median = (numbers_sorted[n//2 - 1] + numbers_sorted[n//2]) / 2
    else:
        median = numbers_sorted[n//2]

    return mean, median

# Example usage
data = [10, 15, 20, 25, 30]
mean, median = calculate_mean_median(data)
print(f"Mean: {mean}, Median: {median}")
```

```python
# Hard Challenge##Simulate a mini inventory system using nested dictionaries and functions
inventory = {
    "Laptops": {
        "Dell": {"quantity": 10, "price": 750},
        "HP": {"quantity": 5, "price": 820}
    },
    "Phones": {
        "iPhone": {"quantity": 7, "price": 999},
        "Samsung": {"quantity": 12, "price": 650}
    }
}

def display_inventory():
    print("Current Inventory:")
    for category, items in inventory.items():
        print(f"\n{category}:")
        for item, details in items.items():
            print(f"  {item} - Qty: {details['quantity']}, Price: ${details['price']}")

def update_quantity(category, item, amount):
    if category in inventory and item in inventory[category]:
        inventory[category][item]['quantity'] += amount
        print(f"Updated {item} quantity to {inventory[category][item]['quantity']}")
    else:
        print("Item not found in inventory.")

def add_new_item(category, item, quantity, price):
    if category not in inventory:
        inventory[category] = {}
    inventory[category][item] = {"quantity": quantity, "price": price}
    print(f"Added {item} to {category} category.")

# Example usage
display_inventory()
update_quantity("Laptops", "Dell", -2)
add_new_item("Tablets", "iPad", 4, 499)
display_inventory()
```
